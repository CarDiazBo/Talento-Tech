```python
# programa de calcular Distancias
from pylab import *
import numpy as np
import matplotlib.pyplot as plt

#valores de las coord
x1=float(input("Cual es la coordenada x1?"))
x2=float(input("Cual es la coordenada x2?"))
y1=float(input("Cual es la coordenada y1?"))
y2=float(input("Cual es la coordenada y2?"))

def dist_euclidiana(x1, x2,y1,y2):
  print("distancia Eucludiana=", np.sqrt(((x2-x1)**2)+((y2-y1)**2)))

def dist_manhattan(x1, x2,y1,y2):
  print("distancia manhattan=", abs(x2-x1)+abs(y2-y1))

def dist_chebyshev(x1, x2,y1,y2):
  print("distancia chebyshev=", max(abs(x2-x1),abs(y2-y1)))

def dist_minkowski(x1, x2,y1,y2,p):
  print("distancia minkowski=", (abs(x2-x1)**p+abs(y2-y1)**p)**(1/p))

def dist_coseno(x1, x2,y1,y2):
  print("distancia coseno=",
        np.arccos(np.dot((x2-x1),(y2-y1))/(np.linalg.norm(x2-x1)*np.linalg.norm(y2-y1))))
```

```python
dist_euclidiana(x1, x2,y1,y2)
dist_manhattan(x1, x2,y1,y2)
dist_chebyshev(x1, x2,y1,y2)
dist_minkowski(x1, x2,y1,y2,2)
dist_coseno(x1, x2,y1,y2)
```
