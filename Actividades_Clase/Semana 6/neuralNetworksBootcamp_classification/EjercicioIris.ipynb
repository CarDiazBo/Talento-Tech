```python
from sklearn import datasets
from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline

iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

promedios = df.mean()

df.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()

df.plot(kind='hist', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()

X = iris.data

# Calculate the mean of the iris dataset
mean_iris = np.mean(X, axis=0)

# Calculate the Euclidean distance to the mean
euclidean_dist = euclidean_distances([mean_iris], X)

# Calculate the Manhattan distance to the mean
manhattan_dist = manhattan_distances([mean_iris], X)

print("Euclidean distances to the mean:\n", euclidean_dist)
print("\nManhattan distances to the mean:\n", manhattan_dist)

# For Euclidean distance
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=euclidean_dist.flatten(), cmap='viridis')
plt.colorbar(label='Euclidean Distance to Mean')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Euclidean Distance to Mean')
plt.show()

# For Manhattan distance
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=manhattan_dist.flatten(), cmap='viridis')
plt.colorbar(label='Manhattan Distance to Mean')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Manhattan Distance to Mean')
plt.show()
```
