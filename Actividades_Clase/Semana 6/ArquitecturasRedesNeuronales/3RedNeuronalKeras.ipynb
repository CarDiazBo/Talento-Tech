```python
import keras as k
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# Genera conjunto de datos de clasificación sintetico
X, y = make_classification(n_samples=100, n_features=5, n_classes=2)
# Dividir los datos en entrenamiento y prueba
X_trainm, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Convertir la etiquetas a codificación one-hot
X_train = k.utils.to_categorical(y_train)
X_test = k.utils.to_categorical(y_test)
```

```python
# Crear una instancia de un modelo secuencial
model = k.models.Sequential()
# Añanir capas al modelo
model.add(k.layers.Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
# Añadir una capa de salida a dos neuronas
model.add(k.layers.Dense(2, activation='softmax'))
```

```python
# Compilar el modelo
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# Entrenar el modelo
y_train_categorical = k.utils.to_categorical(y_train)
y_test_categorical = k.utils.to_categorical(y_test)
model.fit(X_train, y_train_categorical, epochs=10, batch_size=32, validation_data=(X_test, y_test_categorical))
# Evaluación del modelo
loss, accuracy = model.evaluate(X_test, y_test_categorical)
# Impresión del resultado
print('Perdida en el conjunto de prueba:', loss)
print('Precisión en el conjunto de prueba:', accuracy)
```
