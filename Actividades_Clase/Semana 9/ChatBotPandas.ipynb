```python
# Origen Código: https://www.kaggle.com/code/omartronco/chatbot-b-sico-con-menos-de-10-l-neas-de-c-digo/notebook
import pandas as pd

# Se importa una base de datos: https://www.kaggle.com/datasets/yapwh1208/chatbot-ai-q-and-a
data = pd.read_csv('AI1.csv')
data.head()
```

```python
#  Se importan las dos librerías adicionales para vectorizar y para medir la distancia entre vectores.
from sklearn.feature_extraction.text import TfidfVectorizer # Para convertir el texto a vectores
from scipy import spatial # Para medir la distancia entre vectores

vectorizer = TfidfVectorizer()
# Conviertir la columna 'Question' en vectores
X = vectorizer.fit_transform(data.Question) # Cada pregunta la convierte en una serie de números

# El objeto vectorizer ya fue entrenado con la base de datos (tomando en cuenta solo las preguntas). Lo primero que hace es identificar todas las palabras que se encuentren en la base
# Use get_feature_names_out() instead of get_feature_names()
pd.DataFrame(vectorizer.get_feature_names_out())
```

```python
pd.DataFrame(X.toarray()[0]) # Muestra el vector de la primera pregunta
```

```python
question = ['Who created the Hebbian learning rule?']
question

vectorized_question = vectorizer.transform(question).toarray()
pd.DataFrame(vectorized_question)
```

```python
#  identificar cuál es la pregunta de la base de datos, que más se parezca a la pregunta que le hicimos al bot. Debido a que ya todo son vectores, es relativamente fácil medir la distancia entre vectores.
tree = spatial.KDTree(X.toarray())
pd.DataFrame(tree.query(vectorized_question))
# La ditancia que existe entre la pregunta realizada y la pregunta de la base de datos más parecida.
# La respuesta de la base de datos más parecida.
```

```python
# Antes de responder, vemos cuál es la pregunta 0
data.iloc[tree.query(vectorized_question)[1][0]].Question
```

```python
# con esto el bot está listo para responder la pregunta y lanzar una respuesta
data.iloc[tree.query(vectorized_question)[1][0]].Answer
```

```python
# Inicia el chatbot
while True:
    user_input = input("Tú: ")
    if user_input == "salir":
        break
    # Transform user input into a list
    vectorized_question = vectorizer.transform([user_input]).toarray() # Pass user_input as a single-element list
    tree = spatial.KDTree(X.toarray())
    data.iloc[tree.query(vectorized_question)[1][0]].Question
    response = data.iloc[tree.query(vectorized_question)[1][0]].Answer
    print("Chatbot:", response)
```
