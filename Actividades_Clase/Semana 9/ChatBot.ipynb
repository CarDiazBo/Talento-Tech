```python
from deeppavlov import build_model, configs
from deeppavlov.core.data.simple_vocab import SimpleVocabulary

# Load the pre-trained model using the correct configuration name
model_config = configs.faq['fasttext_logreg']
model = build_model(model_config, download=True)

# Access the second component (index 1) and make sure it's not a tuple
component_to_configure = model.pipe[1]
if isinstance(component_to_configure, tuple):
    # If it's a tuple, extract the actual component (assuming it's the SECOND element at index 1)
    component_to_configure = component_to_configure[0] # Changed index from 0 to 1
```

```python
# Inicia el chatbot
while True:
    user_input = input("Tú: ")
    if user_input == "salir":
        break
    response = model([user_input])[0]
    print("Chatbot:", response)
```
