```python
# Gr√°fica de la serie original
plt.figure(figsize=(10, 6))
plt.subplot(411)
plt.plot(ts, label='Serie Original')
plt.title('Serie Original')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.show()

# Gr√°fica de tendencia
plt.subplot(412)
plt.plot(decomposition.trend, label='Tendencia')
plt.title('Tendencia')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()

# Gr√°fica de estacionalidad
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Estacionalidad')
plt.title('Estacionalidad')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()

# Gr√°fica de residuo
plt.subplot(414)
plt.plot(decomposition.resid, label='Residuo')
plt.title('Residuo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()

plt.tight_layout()
plt.show()
```
