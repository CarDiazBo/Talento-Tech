```python
#programa que lee una cadena de y devuelva un diccionario con la cantidad de apariciones de cada palabra en la cadena.
dict = {}
frase = input("Frase:")
lista_palabras=frase.split(" ")
for palabra in lista_palabras:
	if palabra in dict:
		dict[palabra]+=1
	else:
		dict[palabra]=1

for campo,valor in dict.items():
	print (campo,"->",valor)
```

```python
#Programa que verifica si el numero digitado el primo
num=int(input("Número:"))
primo = True
for cont in range(2,num):
    if num%cont==0:
        primo=False
        break
if primo:
    print("Es primo")
else:
    print("No es primo")
print("Fin del programa 2")
```
