```python
import cv2
from google.colab.patches import cv2_imshow  # Import the Colab-compatible function

# Load an image from file
image = cv2.imread('image.jpg')

# Display the image using cv2_imshow
cv2_imshow(image)  # Use the provided alternative for displaying images

# Wait for a key press (this might not be necessary with cv2_imshow)
cv2.waitKey(0)

# Close all open windows (might not be necessary with cv2_imshow)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
#cargar imagen
image = cv2.imread('image.jpg')
# Obtener dimensiones de la escala
height, width = image.shape[:2]
# Calcular el centro de la imagen
center = (width // 2, height // 2)
# Crear una matriz de rotacin
angle = 45
m = cv2.getRotationMatrix2D(center, angle, 1.0)
# Aplicar la rotacin a la imagen
rotated_image = cv2.warpAffine(image, m, (width, height))
# Mostrar la imagen rotada
cv2_imshow(rotated_image) # Remove the window name argument
cv2.waitKey(0)
cv2.destroyAllWindows() # Add parentheses to call the function
```

```python
# Definir la matriz de translacin
tx, ty = 100, 50
M = np.float32([[1, 0, tx], [0, 1, ty]])
# Aplicar la translacin a la imagen
translated_image = cv2.warpAffine(image, M, (width, height))
# Mostrar la imagen transladada
cv2_imshow(translated_image)
# Aplicar la translacin a la imagen
```

```python
# Definir nueva anchura y altura
new_width, new_height = 300, 200

# reescalar la imagen
resized_image = cv2.resize(image, (new_width, new_height))

# Mostrar la imagen reescalada
cv2_imshow(resized_image)
```

```python
# aplicar el filtro Gaussiano para suavizar la imagen
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Mostrar la imagen suavizada
cv2_imshow(blurred_image)
#
```

```python
# Definir el kernel para el filtro de afilado
kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
# Aplicar el filtro del afilado para realzar los detalles
sharpened_image = cv2.filter2D(image, -1, kernel)

# Mostrar la imagen realzada
cv2_imshow(sharpened_image)
#
```

```python
# Cargar la imagen en una escala de grises
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE) # Use IMREAD_GRAYSCALE instead of IMREAD_GRAY
# Aplicar el operador Sobel para detectar bordes
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
# Combinar las respuestas en magnitud
edges = cv2.magnitude(sobelx, sobely)
# Normalizar los valores para ver la imagen correctamente
edges = cv2.normalize(edges, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
# Mostrar laimagen con los bordes
cv2_imshow(sobelx)
cv2_imshow(edges)
```
