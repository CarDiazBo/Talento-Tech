```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
FaceClassif = cv2.CascadeClassifier('/content/haarcascade_frontalcatface.xml')
# Cargar la imagen
image = cv2.imread('Gatos4.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Detección rostros de gatos
faces = FaceClassif.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30,30), maxSize=(200,200))
print(faces)
print(f'Número de rostros detectados: {len(faces)}')
```

```python
# Dibujar rectángulos alrededor de los rostros detectados
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), 2)
```
