```python
import skimage

# Ruta de la imagen
image_path = 'image.jpg'

# Carga la imagen
image = skimage.io.imread(image_path)

# Explora la imagen
print(type(image))
print(image.shape)

# Muestra la imagen
skimage.io.imshow(image)
skimage.io.show
```

```python
# Rotar la imagen
rotated_image = skimage.transform.rotate(image, 30)

# Muestra la imagen rotada
skimage.io.imshow(rotated_image)
skimage.io.show
```

```python
# Aplicar un filtro gaussiano
smoothed_image = skimage.filters.gaussian(image, sigma=2)

# Muestra la imagen filtrada
skimage.io.imshow(smoothed_image)
skimage.io.show
```
