```python
import numpy as np
import matplotlib.pyplot as plt

# Define las listas X y h
X = [1, 2, 3, 4, 5]
h = [6, 7, 8, 9, 10]

# Calcula la convolución
result = np.convolve(X, h)

# Grafica el resultado
plt.plot(result)
plt.title('Convolución de X y h')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def convolve_and_plot(X, h, n, m):
  # Asigna los valores de n y m a las listas
  X[1] = n
  h[2] = m

  # Calcula la convolución
  result = np.convolve(X, h)

  # Grafica el resultado
  plt.plot(result)
  plt.title('Convolución de X y h')
  plt.xlabel('Índice')
  plt.ylabel('Valor')
  plt.show()

# Define las listas iniciales
X = [1, 0, 3, 4, 5]  # n se asignará en la posición 1
h = [6, 7, 0, 9, 10] # m se asignará en la posición 2

# Llama a la función con valores específicos para n y m
convolve_and_plot(X, h, 2, 3)
```

```python
# Convolución a mano
def convolucion(X, h):
    # se establece el tamaño de la lista Y de resultado
    y_length = len(X) + len(h) - 1
    y = [0] * y_length

    print(y_length)
    print(y)

    for i in range(len(X)):
        for j in range(len(h)):
            y[i + j] += X[i] * h[j]

    return y

X = [1, 1, 1, 1, 1]
h = [1, 1, 1, 1, 1]

y = convolucion(X, h)
print(y)

# Grafica el resultado y
plt.plot(y)
plt.title('Convolución de X y h')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.show()
```
